```python
def canonicalize(string):
    normalized_tokens = list()
    a = word_tokenize(string, engine = 'newmm')
    for j in a:
        token = j.strip()
        #Add clean statement here 
        if len(token) > 1 and token not in set(punctuation) and token not in ['..','...','ๆๆ']:
            try:
                normalized_tokens.append(token.lower())
            except:
                normalized_tokens.append(token)
                pass
    return normalized_tokens

print('OK !')
```

```python
# We pass in our tokenizer to the vectorizer object.
tf_vectorizer = CountVectorizer(tokenizer=canonicalize)
tfidf_vectorizer = TfidfVectorizer(tokenizer=canonicalize, sublinear_tf=True)
print('OK')
```

```python
data = pd.read_csv('data/clean_sample_text.csv')
```

```python
print_counts(data)
```

```python
data.shape

data = reset(data[data['content'].notna()])

data.shape
```

```python
start = time.time()

tfidf_term_document_matrix = tfidf_vectorizer.fit_transform(data['content'])

print("Took " + "%.2f" % float(time.time() - start) + " sec")
```

```python
start = time.time()

tf_term_document_matrix = tf_vectorizer.fit_transform(data['content'])

print("Took " + "%.2f" % float(time.time() - start) + " sec")
```

```python
# Train cluster

start = time.time()

document_matrix = tfidf_term_document_matrix

num_clusters = 30

kmeans = KMeans(n_clusters=num_clusters, n_init=2, verbose=10, random_state = 1234)
kmeans.fit(document_matrix)

print("Took " + "%.2f" % float(time.time() - start) + " sec")
```

```python
#General
import collections

text = data['content']

clus_dict = {}
stat_list = []
clus_dict_member = {}
indices_dict = {}

# Group the posts by their cluster labels.
clustering = collections.defaultdict(list)
for idx, label in enumerate(kmeans.labels_):
    clustering[label].append(idx)


for cluster, indices in clustering.items():
    print("\nCluster:", cluster, " Num posts: ", len(indices),' Percent:', round(len(indices)*100/len(text), 2))
    cur_docs = 0
    clus_dict[cluster] = round(len(indices)*100/len(text), 2)
    clus_dict_member[cluster] = len(indices)
    stat_list.append(len(indices)*100/len(text))
    indices_dict[cluster] = indices
```

```python
import operator

sorted_x = sorted(clus_dict.items(), key=operator.itemgetter(1))
sorted_x = sorted_x[::-1]
```

```python
sameple_post_no = 20
sample_post_len = 150
```

```python
for cluster, percent in sorted_x:
    cur_docs = 0
    print("\nCluster:", cluster, " Num posts: ", clus_dict_member[cluster],' Percent:', percent)
    for index in indices_dict[cluster]:
        if (cur_docs > sameple_post_no):
            break
        post_contents = text.iloc[index].replace('\n', '')
        print(index, (post_contents[:sample_post_len] + '..') if len(post_contents) > sample_post_len else post_contents)
        cur_docs+=1
```

```python
# Nearest Posts to the centroids
km = kmeans
X = document_matrix

text = data['content']
nearest_points = sameple_post_no

for j, percent in sorted_x:
    d = km.transform(X)[:, j]
    ind = np.argsort(d)[:nearest_points]
    
    print('Cluster:', j, ' Percent:',clus_dict[j])
    for i in ind:
        print(i, text.iloc[i])
    print()
    print()
    print('___________________________________________________________________________________________')
    print('___________________________________________________________________________________________')    
    print('___________________________________________________________________________________________')  
    print()
    print()
```
